# Weather Analysis of Selected Countries

```python
#import required libraries and dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
import datetime
import scipy.stats as sts
import requests as requests

#import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls

```python
#get base URL
url = "http://api.openweathermap.org/data/2.5/weather?"

#print the massage
print("Beginning Data Retrieval")
print("----------------------------")

#open output file and write the headers
out_file = open(output_data_file,"w")
print(f"City,Cloudiness,Country,Date,Humidity,Lat,Lng,Max Temp,Wind Speed",file = out_file)

#request the weather information for each city
city_num = 1
for selected_city in cities:
    query_url = url + "appid=" + weather_api_key + "&q=" + selected_city
    response = requests.get(query_url)
    response_json = response.json()
    
    try:
        city_name = response_json["name"]
    except:
        print("City not found. Skipping...")
        continue
        
    print(f"Processing Record {city_num} of Set 1 | {city_name}")
    city_num+=1
    
    cloudiness = response_json["clouds"]["all"]
    country = response_json["sys"]["country"]
    date = response_json["dt"]
    humidity = response_json["main"]["humidity"]
    lat = response_json["coord"]["lat"]
    lng = response_json["coord"]["lon"]
    mx_temp = response_json["main"]["temp_max"]
    wind_speed = response_json["wind"]["speed"]
    
    print(f"{city_name},{cloudiness},{country},{date},{humidity},{lat},{lng},{mx_temp},{wind_speed}",file = out_file)

#print the end massage
print("----------------------------")
print("Data Retrieval Complete")
print("----------------------------")

#close the output file
out_file.close()
```

## Convert Raw Data to DataFrame

```python
#generate weather data frame
weather_df = pd.read_csv(output_data_file,delimiter = ",")

#convert the temperature from K to F
weather_df["Max Temp"]=(weather_df["Max Temp"]-273.15)*9/5+32
weather_df.head()
```

## Plotting the Data

```python
#get the date of the data
cur_time = time.ctime(int(weather_df.iloc[0,3]))
time_date=datetime.datetime.strptime(cur_time, '%a %b %d %H:%M:%S %Y').strftime('%m/%d/%Y')

#plot the Max Temp data
weather_df.plot.scatter(x = "Lat",y="Max Temp",edgecolors="Black",s=40,grid = True)
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title(f"City Latiture vs Max Temperature ({time_date})")
plt.savefig("../Results/Lat_vs_MaxTemp.png")
plt.show()
```

```python
#plot the humidity data
weather_df.plot.scatter(x="Lat",y="Humidity",edgecolors="Black",s=40,grid = True)
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.title(f"City Latiture vs Humidity ({time_date})")
plt.savefig("../Results/Lat_vs_Humidity.png")
plt.show()
```

```python
#plot the Cloudiness data
weather_df.plot.scatter(x="Lat",y="Cloudiness",edgecolors="Black",s=40,grid = True)
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.title(f"City Latiture vs Cloudiness ({time_date})")
plt.savefig("../Results/Lat_vs_Cloudiness.png")
plt.show()
```

```python
#plot the Wind speed data
weather_df.plot.scatter(x="Lat",y="Wind Speed",edgecolors="Black",s=40,grid = True)
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.title(f"City Latiture vs Wind Speed ({time_date})")
plt.savefig("../Results/Lat_vs_WindSpeed.png")
plt.show()
```

## Linear Regression

```python
#get north and south hemisphare latitude
n_hemi = weather_df.loc[weather_df["Lat"] >= 0,"Lat"]
s_hemi = weather_df.loc[weather_df["Lat"] < 0,"Lat"]

n_mxTemp = weather_df.loc[weather_df["Lat"] >= 0,"Max Temp"]
s_mxTemp = weather_df.loc[weather_df["Lat"] < 0,"Max Temp"]

n_hum = weather_df.loc[weather_df["Lat"] >= 0,"Humidity"]
s_hum = weather_df.loc[weather_df["Lat"] < 0,"Humidity"]

n_cloud = weather_df.loc[weather_df["Lat"] >= 0,"Cloudiness"]
s_cloud = weather_df.loc[weather_df["Lat"] < 0,"Cloudiness"]

n_wspd = weather_df.loc[weather_df["Lat"] >= 0,"Wind Speed"]
s_wspd = weather_df.loc[weather_df["Lat"] < 0,"Wind Speed"]
```

```python
#Data fitting
(slope, intercept, rvalue,_,_) = sts.linregress(n_hemi, n_mxTemp)
regress_values = n_hemi * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(n_hemi,n_mxTemp)
plt.plot(n_hemi,regress_values,"r-")
plt.annotate(line_eq,(0,0),fontsize=15,color="red")
print("The r-squared is:{:.4f}".format(rvalue))
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title(f"City Latiture vs Max Temperature ({time_date})")
plt.savefig("../Results/Lat_vs_MaxTemp_N_regress.png")
plt.show()
```
